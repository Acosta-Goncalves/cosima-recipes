# Plot sea surface height

This notebook is an example of how to:

- load sea surface height from two different resolution runs
- plot the time mean and std SSH
- compare to plots of AVISO observational data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cosima_cookbook as cc
import numpy as np

import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cft
from dask.distributed import Client
import cmocean as cm
```

```python
client = Client(n_workers=4)
client
```

## Start a Cosima cookbook database session

```python
session = cc.database.create_session()
```

## Load SSH for a 1$^{\circ}$ run

Here we can specify the rough start and end times using the "start_time" and "end_time" arguments. But, this yields a dataset that has an extra 12 months on either end, so we slice the data to match the AVISO dataset below.

```python
expt = '1deg_jra55v13_iaf_spinup1_B1'
variable = 'sea_level'
ssh1 = cc.querying.getvar(expt,variable,session,
                            start_time='1993-01-16',
                            end_time='2010-12-16')
ssh1 = ssh1.isel(time=slice(12,-12)) # slice to the same time range as AVISO
ssh1
```

## Load SSH for a 0.25$^{\circ}$ run

```python
expt = '025deg_jra55v13_ryf8485_gmredi6'
variable = 'sea_level'
ssh25 = cc.querying.getvar(expt,variable,session,
                            start_time='1993-01-16',
                            end_time='2010-12-16')
ssh25 = ssh25.isel(time=slice(12,-12)) # slice to the same time range as AVISO
ssh25
```

# Load AVISO observational data

Use xarray to load 'zos': the sea surface height variable name. Here we slice off the first three time points so that the data starts in January and spans an even 18 years.

```python
obs = xr.open_dataset('/g/data/hh5/tmp/cosima/observations/original/zos_AVISO_L4_199210-201012.nc')
obs_ssh = obs.zos.isel(time=slice(3,None))
obs_ssh
```

# Make comparison plot

Use cartopy to plot the time mean and standard deviation of both of the model datasets along with AVISO data.

```python
plt.figure(figsize=(20,8))
mlev = np.arange(-1.6,1.6,0.02) # Define the contour levels and spacing
slev = np.arange(0,.3,.01)
    
ax = plt.subplot(2,3,1,projection=ccrs.Robinson(central_longitude=-100))
plt.title('Mean SSH (1$^{\circ}$)')
p1 = ssh1.mean('time').plot.contourf(cmap=cm.cm.balance,levels=mlev,add_colorbar=False,
                                extend='both',transform=ccrs.PlateCarree())

ax = plt.subplot(2,3,2,projection=ccrs.Robinson(central_longitude=-100))
plt.title('Mean SSH (0.25$^{\circ}$)')
p1 = ssh25.mean('time').plot.contourf(cmap=cm.cm.balance,levels=mlev,add_colorbar=False,
                                extend='both',transform=ccrs.PlateCarree())

ax = plt.subplot(2,3,3,projection=ccrs.Robinson(central_longitude=-100))
plt.title('AVISO Mean SSH')
p1 = obs_ssh.mean('time').plot.contourf(cmap=cm.cm.balance,levels=mlev,add_colorbar=False,
                                extend='both',transform=ccrs.PlateCarree())

ax2 = plt.subplot(2,3,4,projection=ccrs.Robinson(central_longitude=-100))
plt.title('SSH Standard Deviation (1$^{\circ}$)')
p2 = ssh1.std('time').plot.contourf(cmap=cm.cm.deep,levels=slev,add_colorbar=False,
                                extend='max',transform=ccrs.PlateCarree())

ax2 = plt.subplot(2,3,5,projection=ccrs.Robinson(central_longitude=-100))
plt.title('SSH Standard Deviation (0.25$^{\circ}$)')
p2 = ssh25.std('time').plot.contourf(cmap=cm.cm.deep,levels=slev,add_colorbar=False,
                                extend='max',transform=ccrs.PlateCarree())

ax2 = plt.subplot(2,3,6,projection=ccrs.Robinson(central_longitude=-100))
plt.title('AVISO SSH Standard Deviation')
p2 = obs_ssh.std('time').plot.contourf(cmap=cm.cm.deep,levels=slev,add_colorbar=False,
                                       extend='max',transform=ccrs.PlateCarree())

# Colorbars
ax1 = plt.axes([0.92,0.57,0.01,0.3])
cb = plt.colorbar(p1,cax=ax1,orientation='vertical')
cb.ax.set_ylabel('Mean SSH (m)')

ax2 = plt.axes([0.92,0.14,0.01,0.3])
cb = plt.colorbar(p2,cax=ax2,orientation='vertical')
cb.ax.set_ylabel('SSH Standard Deviation (m)')
```
